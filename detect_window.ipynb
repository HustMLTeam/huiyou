```python
import numpy as np
import cv2
```

```python
pos = [49, 77, 48, 61],
        [85, 112, 46, 58],
        [121, 148, 45, 56],
        [159, 183, 47, 56],
        [192, 216, 47, 59],
```

```python
tube_pos = np.s_[0:271, 33:70]
window_pos = np.s_[192:216, 47:59]
```

```python
window_pos = [[192, 216, 47, 59],
              [159, 183, 47, 56],
              [121, 148, 45, 56],
              [85, 112, 46, 58],
              [49, 77, 48, 61]]
```

```python
first_avg = None
backgrounds = []

p = 0.1
t = 5

count = 0
speed = 1

cap = cv2.VideoCapture('data/tube.mp4')
while cap.isOpened():
    count += 1
    ret, frame = cap.read()
    frame_blue = frame[:, :, 0]
    tube = frame_blue[0:271, 33:70]
    
    if count == 1:  # 第一帧时初始化背景与蓝色平均值
        first_avg = tube.mean()
        for y_start, y_end, x_start, x_end in window_pos:
            backgrounds.append(frame_blue[y_start:y_end, x_start:x_end])
        continue

    frame_blue = frame_blue + first_avg - tube.mean()
    for i, (y_start, y_end, x_start, x_end) in enumerate(window_pos):
        window = frame_blue[y_start:y_end, x_start:x_end]
        foreground = abs(window - backgrounds[i])
        foreground[foreground<=t] = 0

        if not count % 20:  # 更新背景
            tmp = (1 - p) * backgrounds[i] + p * window
            backgrounds[i][foreground==0] = tmp[foreground==0]

        foreground = foreground.astype('uint8')
        fore_bl = cv2.medianBlur(foreground, 7)
        if not (fore_bl > 10).any():
            break
                
    if 0 < i < 5:
        y_start, y_end, x_start, x_end = window_pos[i-1]
        frame = cv2.rectangle(frame, (x_start, y_start), (x_end, y_end), (0, 0, 255), 2)
    
    cv2.imshow('Frame', frame)
    
    key = cv2.waitKey(speed) & 0xff
    if key == ord('q'):
        break
    elif key == ord('p'):
        cv2.waitKey()
    elif key == ord('c'):
        speed = 40 - speed
        
cap.release()
cv2.destroyAllWindows()
```

```python
def show_im(mul, frame=None, *imgs):
    result = None
    height = imgs[0].shape[0]
    cut = cv2.cvtColor(np.ones((height, 1), dtype='uint8') * 255, cv2.COLOR_GRAY2BGR)
    for img in imgs:
        if img.ndim == 2:
            img = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
            
        if result is None:
            result = img
        else:
            result = np.hstack((result, cut, img))
    result = cv2.resize(result, None, fx=mul, fy=mul)
    
    if frame is not None:
        new_width = result.shape[1]
        new_height = int(frame.shape[0] / frame.shape[1] * new_width)
        new_frame = cv2.resize(frame, (new_width, new_height))
        result = np.vstack((new_frame, result))
        
    cv2.imshow('frame', result)
```
