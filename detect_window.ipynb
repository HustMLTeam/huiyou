```python
import numpy as np
from scipy.signal import convolve2d
import cv2
from collections import deque
```

```python
pos = {'tube': [[0, 271, 33, 70],
                [0, 271, 78, 115]],
       'window1': [[49, 77, 48, 61],
                   [85, 112, 46, 58],
                   [121, 148, 45, 56],
                   [159, 183, 47, 56],
                   [192, 216, 47, 59]],
       'window2': [[5, 33, 92, 105],
                   [40, 68, 90, 103],
                   [76, 105, 89, 101],
                   [113, 141, 89, 101],
                   [150, 177, 90, 102],
                   [185, 210, 92, 104],
                   [219, 244, 95, 107]]
       }
```

```python
def show_im(mul, frame=None, *imgs):
    result = None
    height = imgs[0].shape[0]
    cut = cv2.cvtColor(np.ones((height, 1), dtype='uint8') * 255, cv2.COLOR_GRAY2BGR)
    for img in imgs:
        if img.ndim == 2:
            img = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
            
        if result is None:
            result = img
        else:
            result = np.hstack((result, cut, img))
    result = cv2.resize(result, None, fx=mul, fy=mul)
    
    if frame is not None:
        new_width = result.shape[1]
        new_height = int(frame.shape[0] / frame.shape[1] * new_width)
        new_frame = cv2.resize(frame, (new_width, new_height))
        result = np.vstack((new_frame, result))
        
    cv2.imshow('frame', result)
```

```python
def median_filter(src):
    result = convolve2d(src, np.ones((1, 5))/5, mode='same')
    ret, result = cv2.threshold(np.uint8(result), 200, 255, cv2.THRESH_BINARY)
    return result
```

```python
window_pos = [[7, 31, 94, 103],
              [42, 66, 92, 101],
              [78, 103, 91, 99],
              [115, 139, 91, 99],
              [152, 175, 92, 101],
              [187, 208, 94, 102],
              [221, 242, 97, 105]]

init_level = 160
levels = deque([init_level]*30, maxlen=30)

upper = []
lower = []
for pos in window_pos:
    y_start, y_end, x_start, x_end = pos
    if y_end <= init_level:
        upper.append(pos)
    elif y_start < init_level < y_end:
        upper.append([y_start, init_level, x_start, x_end])
        lower.append([init_level+1, y_end, x_start, x_end])
    else:
        lower.append(pos)

first_avg = None
backgrounds = []
result = None

p = 0.1
t = 3
d = 30

count = 0
speed = 1

cap = cv2.VideoCapture('data/tube.mp4')
while cap.isOpened():
    count += 1
    ret, frame = cap.read()
    frame_blue = frame[:, :, 0]
    tube = frame_blue[0:271, 33:70].astype('float64')
   
    if count == 1:  # 第一帧时初始化背景与蓝色平均值
        first_avg = tube.mean()
        for y_start, y_end, x_start, x_end in reversed(lower):
            backgrounds.append(frame_blue[y_start:y_end, x_start:x_end].astype('float64'))
        for y_start, y_end, x_start, x_end in upper:
            backgrounds.append(frame_blue[y_start:y_end, x_start:x_end].astype('float64'))
        continue
        
    sobels = []
        
    i = 0
    result = []
    white = None
    level_mean = int(np.mean(levels))
    # 在初始线以下寻找
    for y_start, y_end, x_start, x_end in reversed(lower):
        window = frame_blue[y_start:y_end, x_start:x_end].astype('float64')
        window = window + first_avg - tube.mean()
        foreground = abs(window - backgrounds[i])
        foreground[foreground<=t] = 0

        if not count % 20:  # 更新背景
            tmp = (1 - p) * backgrounds[i] + p * window
            backgrounds[i][foreground==0] = tmp[foreground==0]
            
        threshold = 25
        i += 1

        y1 = max(level_mean-d, y_start) - y_start
        y2 = min(level_mean+d, y_end) - y_start
        if y1 < y2 and white is None:
            if foreground.mean() > 15:
                white = y2 + y_start
            else:
                sobel = cv2.Sobel(foreground[y1:y2], cv2.CV_16S, 0, 1)
                sobel = np.where(sobel<-threshold, 255, 0).astype('uint8')
                sobel_bl = median_filter(sobel)
                result += list(np.where(sobel_bl)[0] + y_start)
       
    # 在初始线以上寻找
    for y_start, y_end, x_start, x_end in upper:
        window = frame_blue[y_start:y_end, x_start:x_end].astype('float64')
        window = window + first_avg - tube.mean()
        foreground = abs(window - backgrounds[i])
        foreground[foreground<=t] = 0

        if not count % 20:  # 更新背景
            tmp = (1 - p) * backgrounds[i] + p * window
            backgrounds[i][foreground==0] = tmp[foreground==0]
            
#         threshold = max(backgrounds[i].std() * 2.5, 30)
        threshold = 33
        i += 1
            
        y1 = max(level_mean-d, y_start) - y_start
        y2 = min(level_mean+d, y_end) - y_start
        if y1 < y2 and white is None:
            if foreground.mean() > 18:
                white = y2 + y_start
            else:
                sobel = cv2.Sobel(foreground[y1:y2], cv2.CV_16S, 0, 1)
                sobel = np.where(sobel>threshold, 255, 0).astype('uint8')
                sobel_bl = median_filter(sobel)
                result += list(np.where(sobel_bl)[0] + y_start)


    if result:
        levels.append(np.median(result))
    elif white is not None:
        levels.append(white)
        
    if count < 3000:
        continue

    y = int(np.mean(levels))
    frame = cv2.line(frame, (90, y), (105, y), (0, 0, 255), 1)
    frame = cv2.resize(frame, None, fx = 2, fy = 2)
    
    cv2.imshow("result", frame)

    key = cv2.waitKey(speed) & 0xff
    if key == ord('q'):
        break
    elif key == ord('p'):
        print('pause')
        cv2.waitKey()
    elif key == ord('c'):
        print('change speed')
        speed = 40 - speed
       
cap.release()
cv2.destroyAllWindows()
```

```python
# 第二列第四个窗口
cap = cv2.VideoCapture('data/tube.mp4')

count = 0

p = 0.1
t = 3
speed = 1

while cap.isOpened():
    count += 1
    
    ret, frame = cap.read()
    if not ret:
        print('quit')
        break
        
    tube = frame[:, 33:70, 0]
    window = frame[113:141, 89:101]
    blue = window[:, :, 0]
    
    if count == 1:
        background = blue.astype('float64')
        first_avg = tube.mean()
    else:
        blue = blue + first_avg - tube.mean()
        
    foreground = np.absolute(blue - background).astype('uint8')
    foreground[foreground<=t] = 0
    
    if not count % 20:
        tmp = p * blue + (1 - p) * background
        background[foreground==0] = tmp[foreground==0]
        
#     fore_bl = cv2.medianBlur(foreground, 3)
    sobel_ = cv2.Sobel(foreground, cv2.CV_8U, 0, 1)
    ret, sobel = cv2.threshold(sobel_, 33, 255, cv2.THRESH_BINARY)
    sobel_bl = median_filter(sobel)
    
    if count < 18000:
        continue
#     else:
#         key = cv2.waitKey() & 0xff
    
    show_im(8, frame, window, blue.astype('uint8'), 
            background.astype('uint8'), foreground, sobel, sobel_bl)

    key = cv2.waitKey(speed) & 0xff
    if key == ord('q'):
        print('quit')
        break
    elif key == ord('p'):
        print('pause')
        cv2.waitKey()
    elif key == ord('c'):
        print('change speed')
        speed = 40 - speed
        
cap.release()
cv2.destroyAllWindows()
```

```python
# 在液位仪上画线
cap = cv2.VideoCapture('data/tube.mp4')

tube_start, tube_end = pos['tube'][1][-2:]
window_pos = pos['window2']

q = deque((130 for i in range(10)), maxlen=10)

first_avg = None
background = None
level = np.mean(q)

p = 0.1
t = 2
DIF = 30

count = 0
speed = 1

while cap.isOpened():
    count += 1
    
    ret, frame = cap.read()
    if not ret:
        break
    origin = frame[:, tube_start:tube_end]
    blue = origin[:, :, 0].astype('float64')
    
    if first_avg is None:
        first_avg = blue.mean()
        background = blue
        foreground = np.zeros(blue.shape)
    else:
        blue = blue + first_avg - blue.mean()
        foreground = abs(blue - background)
        foreground[foreground<=t] = 0
        
    if not count % 20:  # 更新背景
        tmp = (1 - p) * background + p * blue
        background[foreground==0] = tmp[foreground==0]

        
    back = background.astype('uint8')
    foreground = foreground.astype('uint8')
    fore_bl = cv2.medianBlur(foreground, 3)
    
    edges = cv2.Canny(fore_bl, 30, 65, L2gradient=True)
    
    detect_area = np.zeros(back.shape, dtype='uint8')
    for y_start, y_end, x_start, x_end in window_pos:
        x_start -= tube_start
        x_end -= tube_start
        detect_area[y_start:y_end, x_start:x_end] = edges[y_start:y_end, x_start:x_end]
    

    
    if np.any(detect_area):
        top = np.where(detect_area>100)[0][0]
        if abs(top - level) < DIF:
            q.append(top)
    else:
        q.append(130)
    level = np.mean(q)
            
#     if count < 10000:
#         continue
        
    result = cv2.line(frame, (tube_start, int(level)), (tube_end, int(level)), (0, 0, 255), 1)
#     result = cv2.resize(result, None, fx=2, fy=2)
#     cv2.imshow('Result', result)
    show_im(1.5, result, origin, blue.astype('uint8'), back, foreground, fore_bl, detect_area, 
             edges)
    key = cv2.waitKey(speed) & 0xff
    if key == ord('q'):
        print('quit')
        break
    elif key == ord('p'):
        print('pause')
        cv2.waitKey()
    elif key == ord('c'):
        print('change speed')
        speed = 40 - speed
        
cap.release()
cv2.destroyAllWindows()
```
